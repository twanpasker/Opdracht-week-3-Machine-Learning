# Introductie
Ik heb voor de voice dataset gekozen omdat deze dataset mij het meest interessant leek om te gebruiken.

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
#importeren van libraries
from sklearn.model_selection import train_test_split #importen van deze library voor het splitten van data
```

```python
df = pd.read_csv("voice.csv") #inladen van het voice csv bestand
df.head(30) #kijken naar variabelen die handig zijn
```

# Data cleaning
Na de dataset hierboven en in Kaggle bekeken te hebben lijken de variabelen 'meanfreq', 'sd', 'median', 'IQR', 'mode' en 'label' mij het beste om mee te werken. Deze variabelen zijn namelijk getallen en sluiten het beste aan op de belangrijkste variabele van "label".

```python
df = df[['meanfreq', 'sd', 'median', 'IQR', 'mode', 'label']]
df = df.dropna() #rijen met NaN weghalen zodat het K-NN algoritme werkt
df.head()
df['label'].value_counts() #kijken hoeveel vrouwen en mannen er in de dataset staan 
```

```python
#hieronder zet ik de kolom "label" om naar getallen
df['label'] = df['label'].replace(['female', 'male'], ['0', '1']) #label = 0 betekent dat het label female is, label = 1 is het label male
df
```

# Predictive model
Ik heb voor het k-NN algoritme gekozen. K-NN staat voor k-nearest neighbor en is een van de meest simpele algoritmes. Met dit algoritme kan je verschillende clusters vormen uit data op basis van de dichtsbij gelegen datapunten.

```python
X = df[['meanfreq', 'sd', 'median', 'IQR', 'mode']] #X-variabele aanmaken

y = df['label'] #Y-variabele aanmaken
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #de data splitten in een training- en testset.
X_train.head() #de eerste rijen van de trainingset bekijken
```

```python
from sklearn.neighbors import KNeighborsClassifier #libraby importeren

knn = KNeighborsClassifier() #Het aanmaken van een k-NN classifier die clusters met 5 datapunten die dichtbij staan herkent
knn = knn.fit(X_train, y_train) #het model trainen met de trainingsset
```

# Evaluation

# Accurancy
Accurancy: welke proportie is goed voorspeld?

```python
knn.score(X_test, y_test)
```

De accurancy van het model is 94,8%. Dit houdt in 94,8% van alle vrouwen en mannen de label goed is voorspeld.
Ik vind 94,8% hoog, want door het hoge percentage kan het model betrouwbare voorspellingen doen.

```python
from sklearn.metrics import confusion_matrix #importeren van confusion matrix library
y_test_pred = knn.predict(X_test) #de voorspelde waardes
cm = confusion_matrix(y_test, y_test_pred) #creeeren van een "confusion matrix"
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['vrouw (actual)', 'man (actual)'], columns = ['vrouw (predicted)', 'man (predicted)']) #confusion matrix aanmaken
conf_matrix 
```
